```python
# Ejercicio 1

import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, lambdify, init_printing

init_printing()  # Habilitar LaTeX
x = symbols('x')  # Definir la variable y la función simbólica
funcion = 2 * x**2 - 4 * x + 1
integrando = integrate(funcion, x)  # Calcular la integral indefinida
display(funcion, integrando)  # Mostrar la función original y su integral simbólica

# Convertir las funciones simbólicas en funciones numéricas
funcion_num = lambdify(x, funcion, 'numpy')
integral_funcion_num = lambdify(x, integrando, 'numpy')

# Crear un rango de valores para graficar
x_vals = np.linspace(-10, 10, 400)
y_vals = funcion_num(x_vals)
y_integral_vals = integral_funcion_num(x_vals)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$f(x) = x^2 + 3x$', color='blue')
plt.plot(x_vals, y_integral_vals,
         label=r'$\int f(x) \,dx$', color='green', linestyle='--')

# Etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.title('Función y su Integral')
plt.legend()
plt.grid(True)
plt.show()


# Ejercicio 2

import matplotlib.pyplot as plt
import numpy as np
# Define la función f(x) = e^(0.5x)
def f(x):
    return np.exp(0.5 * x)
# Crea un arreglo de valores para x en el rango [-2, 5]
x = np.linspace(-2, 5, 100)
# Calcula los valores de f(x) y su integral indefinida
y = f(x)
integral = np.exp(0.5 * x) / 0.5 # La integral de e^(0.5x) es e^(0.5x) / 0.5
# Crea la gráfica de la función y su integral
plt.plot(x, y, label='f(x)')
plt.plot(x, integral, label='Integral de f(x)')
# Agrega etiquetas a los ejes
plt.xlabel('x')
plt.ylabel('y')
# Agrega un título a la gráfica
plt.title('Función Exponencial y su Integral')
# Muestra la leyenda de la gráfica
plt.legend()
# Muestra la gráfica
plt.show()

import numpy as np
import matplotlib.pyplot as plt

# Ejercicio 3: Función Trigonométrica
def f_trig(x):
  return np.sin(x) + np.cos(2*x)

def integral_trig(x):
  return -np.cos(x) + 0.5*np.sin(2*x)

x = np.linspace(-np.pi, np.pi, 100)
y_trig = f_trig(x)
y_integral = integral_trig(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y_trig, label="f(x)")
plt.plot(x, y_integral, label="Integral de f(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ejercicio 3: Función Trigonométrica")
plt.legend()
plt.grid(True)
plt.show()

# Ejercicio 4: Función Racional
def f_racional(x):
  return 1/(x**2 + 1)

def integral_racional(x):
  return np.arctan(x)

x = np.linspace(-10, 10, 100)
y_racional = f_racional(x)
y_integral = integral_racional(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y_racional, label="f(x)")
plt.plot(x, y_integral, label="Integral de f(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ejercicio 4: Función Racional")
plt.legend()
plt.grid(True)
plt.show()


# Ejercicio 5: Función Logarítmica

import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = ln(x + 2)
def f(x):
  return np.log(x + 2)

# Define the range of x values
x = np.linspace(-1, 5, 100)

# Calculate the function values
y = f(x)

# Calculate the integral of the function
y_int = np.zeros_like(x)
for i in range(1, len(x)):
  y_int[i] = y_int[i-1] + (x[i] - x[i-1]) * y[i]

# Plot the function and its integral
plt.plot(x, y, label='f(x)')
plt.plot(x, y_int, label='Integral of f(x)')

# Set the plot title and axis labels
plt.title('Funcion y su Integral')
plt.xlabel('x')
plt.ylabel('y')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
