```python
#PARTE 1
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, diff, lambdify, init_printing, exp, cos, sin, log

init_printing()  # Habilitar LaTeX
x = symbols('x')  # Definir la variable simbólica

# Lista de funciones para integrar
funciones = [
    3*x**2 + 2,
    5*x**4 - 2*x**3,
    exp(x**2),  # Usar exp de SymPy
    cos(x) * sin(x),  # Usar cos y sin de SymPy
    x / (1 + x**2),
    (2*x + 1) * exp(x**2 + x),  # Usar exp de SymPy
    x * cos(x**2),  # Usar cos de SymPy
    1 / (x * log(x))  # Usar log de SymPy
]

# Calcular integrales simbólicas y derivadas
for funcion in funciones:
    integral = integrate(funcion, x)
    derivada = diff(integral, x)
    print(f"Función: {funcion}")
    print(f"Integral: {integral}")
    print(f"Derivada de la integral: {derivada}")
    print("-" * 40)
#PARTE 2
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, diff, lambdify, init_printing, exp, sin, sec, cos

init_printing()  # Habilitar LaTeX
x = symbols('x')  # Definir la variable simbólica

# Lista de funciones para integrar
funciones_segunda_parte = [
    2*x + 3,
    exp(3*x),  # Usar exp de SymPy
    sin(2*x),  # Usar sin de SymPy
    1 / x,
    sec(x)**2,  # Usar sec de SymPy
    cos(3*x)  # Usar cos de SymPy
]

# Calcular integrales simbólicas y derivadas
for funcion in funciones_segunda_parte:
    integral = integrate(funcion, x)
    derivada = diff(integral, x)
    print(f"Función: {funcion}")
    print(f"Integral: {integral}")
    print(f"Derivada de la integral: {derivada}")
    print("-" * 40)
#PARTE 3
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, lambdify, exp, sin

# Definir la variable simbólica
x = symbols('x')

# Funciones y sus intervalos
funciones_area = [
    (x**2, (1, 3)),
    (4 - x**2, (0, 2)),
    (exp(x), (0, 1)),  # Usar exp de SymPy
    (x, (0, 2)),
    (sin(x), (0, np.pi)),  # Usar sin de SymPy
    (1 / x, (1, 2))
]

# Calcular áreas y graficar
for funcion, intervalo in funciones_area:
    a, b = intervalo
    area = integrate(funcion, (x, a, b))
    print(f"Función: {funcion}")
    print(f"Intervalo: [{a}, {b}]")
    print(f"Área: {area}")
    print("-" * 40)

    # Convertir la función simbólica en una función numérica
    funcion_num = lambdify(x, funcion, 'numpy')

    # Crear un rango de valores para graficar
    x_vals = np.linspace(a, b, 400)
    y_vals = funcion_num(x_vals)

    # Crear la gráfica
    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, y_vals, label=f'$f(x) = {funcion}$', color='blue')
    plt.fill_between(x_vals, y_vals, where=(x_vals >= a) & (x_vals <= b), color='green', alpha=0.3)

    # Etiquetas y leyenda
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Área bajo la curva de $f(x) = {funcion}$ en [{a}, {b}]')
    plt.legend()
    plt.grid(True)
    plt.show()
```
